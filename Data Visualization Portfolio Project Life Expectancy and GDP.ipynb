```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data=pd.read_csv("all_data.csv")
print(data.head())
print(data["Country"].unique())
```

```python
Chile=data[data.Country=="Chile"].reset_index()
print(Chile)
```

```python
China=data[data.Country=="China"].reset_index()
print(China)
```

```python
Germany=data[data.Country=="Germany"].reset_index()
print(Germany)
```

```python
Mexico=data[data.Country=="Mexico"].reset_index()
print(Mexico)
```

```python
USA=data[data.Country=="United States of America"].reset_index()
print(USA)
```

```python
Zimbabwe=data[data.Country=="Zimbabwe"].reset_index()
print(Zimbabwe)
```

```python
ax=plt.subplot()
plt.plot(Chile["Year"],Chile["GDP"])
plt.plot(China["Year"],China["GDP"])
plt.plot(Germany["Year"],Germany["GDP"])
plt.plot(Mexico["Year"],Mexico["GDP"])
plt.plot(USA["Year"],USA["GDP"])
plt.plot(Zimbabwe["Year"],Zimbabwe["GDP"])
plt.legend(data["Country"].unique())
plt.xlabel("Year")
plt.ylabel("GDP ($)")
plt.title("Country's GDP along the years")
plt.show()
```

```python
ax=plt.subplot()
plt.plot(Chile["Year"],Chile["Life expectancy at birth (years)"])
plt.plot(China["Year"],China["Life expectancy at birth (years)"])
plt.plot(Germany["Year"],Germany["Life expectancy at birth (years)"])
plt.plot(Mexico["Year"],Mexico["Life expectancy at birth (years)"])
plt.plot(USA["Year"],USA["Life expectancy at birth (years)"])
plt.plot(Zimbabwe["Year"],Zimbabwe["Life expectancy at birth (years)"])
plt.legend(data["Country"].unique())
plt.xlabel("Year")
plt.ylabel("Life expectancy at birth (Years)")
plt.title("Country's life expectancy at birth along the years")
plt.show()
```

```python
ax=plt.subplot()
plt.scatter(data["GDP"],data["Life expectancy at birth (years)"])
plt.xlabel("GDP($)")
plt.ylabel("Life expectancy at birth (Years)")
plt.title("Country's life expectancy VS Country's GDP")
plt.show()
```
